```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

1- Crop and flip an image using Numpy array indexing.

```python
img = cv2.cvtColor(cv2.imread('../dataset/img.jpg'), code=cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# Crop image
img_crop = np.array(img)[0:250, 0:500]
plt.imshow(img_crop)
plt.show()
```

```python
# Crop and flip image
img_cf = np.flip(img_crop, axis=1)
plt.imshow(img_cf)
plt.show()
```

2- Implement image translation using Numpy and OpenCV.

```python
# Translation matrix
tx, ty = 100, 100
h, w = img.shape[:2]
t_matrix = np.array([[1, 0, tx], [0, 1, ty]], dtype=np.float32)

img_t = cv2.warpAffine(img, t_matrix, dsize=(w, h))
plt.imshow(img_t)
plt.show()
```

3- Implement image rotation using Numpy and OpenCV.

```python
# Rotation matrix
h, w = img.shape[:2]
r_matrix = cv2.getRotationMatrix2D((w/2, h/2), 10, 1)

img_r = cv2.warpAffine(img, r_matrix, dsize=(w, h))
plt.imshow(img_r)
plt.show()
```

4- Implement image resizing using OpenCV.

```python
img_resize = cv2.resize(img, (400, 200))
plt.imshow(img_resize)
plt.show()
```

5- Implement bitwise operations: AND, OR, XOR.

```python
shape = np.zeros_like(img, np.uint8)
rectangle = cv2.rectangle(shape, (350, 150), (500, 270), (255, 255, 255), cv2.FILLED)
plt.imshow(rectangle)
plt.show()
```

```python
# AND op.
img_and = cv2.bitwise_and(img, rectangle)
plt.imshow(img_and)
plt.show()
```

```python
# OR op.
img_or = cv2.bitwise_or(img, rectangle)
plt.imshow(img_or)
plt.show()
```

```python
# XOR op.
img_xor = cv2.bitwise_xor(img, rectangle)
plt.imshow(img_xor)
plt.show()
```
