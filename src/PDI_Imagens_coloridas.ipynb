## Processamento de imagens coloridas

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def showChannels(image):
    plt.figure(figsize=(15, 25))
    plt.subplot(1,4,1), plt.imshow(image, cmap='gray')
    plt.subplot(1,4,2), plt.imshow(image[:, :, 0], cmap='gray')
    plt.subplot(1,4,3), plt.imshow(image[:, :, 1], cmap='gray')
    plt.subplot(1,4,4), plt.imshow(image[:, :, 2], cmap='gray')
    plt.show()
```

```python
# Baboon
baboon = cv2.imread('baboon.png')
showChannels(cv2.cvtColor(baboon, cv2.COLOR_BGR2RGB))
```

```python
# RGBCube
rgbcube = cv2.imread('rgbcube_kBKG.png')
showChannels(cv2.cvtColor(rgbcube, cv2.COLOR_BGR2RGB))
```

```python
# RGB
rgb = cv2.imread('rgb.png')
showChannels(cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB))
```

```python
def plotModels(image):
    fig, axs = plt.subplots(4, 7)
    fig.set_figheight(15)
    fig.set_figwidth(25)
    titles = ['RGB', 'YCrCb', 'HSV', 'HLS', 'Lab', 'CIE XYZ', 'CMYK']
    models = [cv2.COLOR_BGR2RGB, cv2.COLOR_BGR2YCrCb, cv2.COLOR_BGR2HSV, cv2.COLOR_BGR2HLS, cv2.COLOR_BGR2LAB, cv2.COLOR_BGR2XYZ, cv2.COLOR_BGR2RGB]

    for row in range(4):
        for col, model in enumerate(models):
            img = cv2.cvtColor(image, model)
            if row == 0:
                axs[row, col].imshow(img, cmap='gray'); axs[row, col].set_title(titles[col])
            else: 
                axs[row, col].imshow(img[:, :, row-1], cmap='gray'); axs[row, col].set_title(titles[col])
```

```python
plotModels(baboon)
```

```python
plotModels(rgbcube)
```

```python
plotModels(rgb)
```
